```python
# Importando as bibliotecas necessárias
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, roc_auc_score
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier

pd.set_option('display.max_columns', None)
```

```python
# Importação dos dados e pré-processamento
```

```python
# Este dataset foi baixado de https://www.kaggle.com/datasets/iamsouravbanerjee/heart-attack-prediction-dataset
dataset = './heart_attack_prediction_dataset.csv' 
df = pd.read_csv(dataset, delimiter=',', encoding='ascii')

df.info() # Para verificar os tipos das colunas

# Separar pressão sistólica e diastólica
df[['Systolic', 'Diastolic']] = (
    df['Blood Pressure'].str.split('/', expand=True).astype(float)
)

# Criar coluna com a média
df['Blood Pressure Mean'] = (df['Systolic'] + 2 * df['Diastolic']) / 3 # Separando apressão cardíaca entre sistólica e diastólica

print(df.isnull()) # Verifica nulos
print(df.duplicated()) # Verifica duplicados

df.head()
```

```python
# Análise exploratória
```

```python
df.describe()

# Perguntas
# 1- Qual a distribuição de chances de ataque cardíaco?
# 2- Qual a relação entre idade, gênero e chances de ataque cardíaco?
# 3- Qual a relação entre colesterol e chances de ataque cardíaco?
# 4- Qual a relação entre diabetes e chances de ataque cardíaco?
# 5- Qual a relação entre triglicerídeos e chances de ataque cardíaco?
# 6- Qual a relação entre pressão, frequência cardíaca e chances de ataque cardíaco?
# 7- Fumar, obesidade e consumo de álcool aumentam as chances de ataque cardíaco?
# 8- Exercício físico e dieta influenciam o risco?
# 9- Histórico familiar ou Problemas cardíacos prévios influenciam nas chances de ataque cardíaco?
```

```python
# 1- Qual a distribuição de chances de ataque cardíaco?
ax = sns.countplot(data=df, x='Heart Attack Risk')

for container in ax.containers:
    ax.bar_label(container)

plt.title('Distribuição de Risco de Ataque Cardíaco')
plt.show()
```

```python
# 0 = não há riscos e 1 = há riscos
# Temos o total de pacientes analisados = 8763
# A porcentagem de pacientes sem risco de ataque cardíaco é de 64,18%
# A porcentagem de pacientes com risco de ataque cardíaco é de 35,82%
# 64,18% + 35,82% totaliza os 100%
```

```python
# 2- Qual a relação entre idade, gênero e chances de ataque cardíaco?

plt.figure(figsize=(24, 5)) 

plt.subplot(1, 3, 1)
sns.boxplot(data=df, x='Sex', y='Age', hue='Heart Attack Risk')
plt.title('Correlação entre Idade, Gênero e Risco')

plt.subplot(1, 3, 2)
sns.countplot(data=df, x='Age', hue='Heart Attack Risk')
plt.title('Contagem de casos por Idade e Risco')

plt.subplot(1, 3, 3)
sns.histplot(df['Age'], kde=True)
plt.title('Outliers na Idade')

plt.tight_layout()
plt.show()
```

```python
# Pacientes mais velhos tentem a ter mais riscos
# Para homens, as chances aumentam mais tarde, próximo a 36 anos e diminuem mais tarde, próximo a 73 anos
# Para mulheres, as chances aumentam mais cedo, próximo a 34 anos, e dimiunem também, próximo a 73 anos
# Pacientes mais velhos tentem a ter mais riscos, mas gênero não influencia muito no risco
```

```python
# 3- Qual a relação entre colesterol e chances de ataque cardíaco?
plt.figure(figsize=(24, 5))

plt.subplot(1, 3, 1)
sns.boxplot(data=df, x='Heart Attack Risk', y='Cholesterol')
plt.title('Correlação entre Colesterol e Risco')

plt.subplot(1, 3, 2)
df['Cholesterol Status'] = df['Cholesterol'].apply(lambda x: 'Desejável' if x < 200 else 'Alto')

ax = sns.countplot(data=df, x='Cholesterol Status', hue='Heart Attack Risk')
plt.title('Risco de Ataque Cardíaco por Nível de Colesterol')

for container in ax.containers:
    ax.bar_label(container)

plt.subplot(1, 3, 3)
sns.histplot(df['Cholesterol'], kde=True)
plt.title('Outliers em Colesterol')

plt.tight_layout()
plt.show()
```

```python
# 0 = não há riscos e 1 = há riscos
# A mediana de colesterol para pacientes com risco é um pouco maior
# Os níveis desejáveis de colesterol são abaixo de 200, e é considerado alto acima de 240
# Para pacientes com colesterol alto, as riscos são maiores. São 63,60%
# Para pacientes com colesterol desejável, os riscos diminuem. São 36,40%
```

```python
# 4- Qual a relação entre diabetes e chances de ataque cardíaco?
ax = sns.countplot(data=df, x='Heart Attack Risk', hue='Diabetes')
plt.title('Correlação entre Diabetes e Risco')

for container in ax.containers:
    ax.bar_label(container)

plt.tight_layout()
plt.show()
```

```python
# 0 = não há riscos e 1 = há riscos
# Temos a maioria de pacientes diabéticos nos dois grupos
# No grupo de pacientes com risco, a presença de diabéticos é de 66,36%
# No grupo de pacientes sem risco, a presença de diabéticos é de 64,61%
```

```python
# 5- Qual a relação entre triglicerídeos e chances de ataque cardíaco?
plt.figure(figsize=(24, 5))

plt.subplot(1, 3, 1)
sns.violinplot(data=df, x='Heart Attack Risk', y='Triglycerides', inner='box')
plt.title('Correlação entre Triglicerídeos e Risco')


plt.subplot(1, 3, 2)
def classify_triglycerides(value):
    if value < 150:
        return 'Desejável'
    elif value < 200:
        return 'Alto'
    else:
        return 'Muito Alto'

df['Triglycerides Status'] = df['Triglycerides'].apply(classify_triglycerides)

ax = sns.countplot(data=df, x='Triglycerides Status', hue='Heart Attack Risk')
plt.title('Triglicerídeos x Risco de Ataque Cardíaco')

for container in ax.containers:
    ax.bar_label(container)

plt.subplot(1, 3, 3)
sns.histplot(df['Triglycerides'], kde=True)
plt.title('Outliers em Triglicerídeos')

plt.tight_layout()
plt.show()
```

```python
# 0 = não há riscos e 1 = há riscos.
# A mediana de triglicerídeos para pacientes com risco é levemente maior do que para pacientes sem risco (423 mg/dL vs 414 mg/dL).
# Níveis desejáveis de triglicerídeos são abaixo de 150 mg/dL, altos entre 150 e 199 mg/dL, e muito altos acima ou iguais a 200 mg/dL.
# Para pacientes com triglicerídeos muito altos, a taxa de risco é maior 37,00%, seguida por desejável 36,60% e alto 34,20%
```

```python
# 6- Qual a relação entre pressão, frequência cardíaca e chances de ataque cardíaco?
plt.figure(figsize=(24, 5))

plt.subplot(1, 3, 1)
sns.kdeplot(
    data=df, x='Heart Rate', y='Systolic',
    hue='Heart Attack Risk', fill=True, thresh=0.1, alpha=0.5
)
plt.title('Correlação entre Pressão, Frequência e Risco')

plt.subplot(1, 3, 2)
sns.histplot(df['Systolic'], kde=True)
plt.title('Outliers na pressão Sistólica')

plt.subplot(1, 3, 3)
sns.histplot(df['Heart Rate'], kde=True)
plt.title('Outliers na Frequência Cardíaca')

plt.show()

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.kdeplot(
    data=df, x='Heart Rate', y='Diastolic',
    hue='Heart Attack Risk', fill=True, thresh=0.1, alpha=0.5
)
plt.title('Correlação entre Pressão Diastólica e Risco')

plt.subplot(1, 2, 2)
sns.histplot(df['Diastolic'], kde=True)
plt.title('Outliers na Pressão Diastólica')

plt.tight_layout()
plt.show()



plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.histplot(df['Blood Pressure Mean'], kde=False, bins=20)

plt.subplot(1, 2, 2)
sns.histplot(df['Blood Pressure Mean'], kde=True)
plt.title('Outliers na Média da Pressão Sanguínea')

plt.tight_layout()
plt.show()


```

```python
# 0 = não há riscos e 1 = há riscos
# As distribuições de pressão e frequência são muito semelhantes para pacientes com e sem risco
# A relação entre pressão diastólica e risco também é muito próxima entre os grupos
# No geral, pressão arterial e frequência cardíaca, de forma isolada, não são muito relevantes para a análise
```

```python
# 0 = não há riscos e 1 = há riscos
# As variáveis fumar, obesidade e consumo de álcool apresentam distribuições muito semelhantes entre pacientes com e sem risco
# As correlações dessas variáveis com o risco são próximas de zero
# De forma isolada, esses fatores não se mostraram relevantes para a análise
```

```python
# 8- Exercício físico e dieta influenciam o risco?

plt.figure(figsize=(18, 5))

# Horas de exercício
plt.subplot(1, 3, 1)
sns.boxplot(data=df, x='Heart Attack Risk', y='Exercise Hours Per Week')
plt.title('Correlação entre Horas de exercício por semana e Risco')

# Dieta
plt.subplot(1, 3, 2)
sns.countplot(data=df, x='Diet', hue='Heart Attack Risk')
plt.xticks(rotation=45)
plt.title('Correlação entre Dieta e Risco')

# Dias de atividade física
plt.subplot(1, 3, 3)
sns.boxplot(data=df, x='Heart Attack Risk', y='Physical Activity Days Per Week')
plt.title('Correlação entre Dias de atividade física e Risco')

plt.show()
```

```python
# 0 = não há riscos e 1 = há riscos
# As distribuições para histórico familiar e problemas cardíacos prévios são praticamente iguais entre pacientes com e sem risco
# De forma isolada, essas variáveis não se mostraram relevantes para a análise
```

```python
# Treinamento de modelo de classificação
```

```python
# Features e alvo

# Lista dos atributos relevantes
relevant_cols = [
    'Age', 'Sex', 'Cholesterol', 'Triglycerides', 'Systolic', 'Diastolic', 'Blood Pressure Mean', 'Heart Rate', 'Exercise Hours Per Week', 'Diet', 
    'Physical Activity Days Per Week', 'Diabetes', 'Smoking', 'Obesity', 'Alcohol Consumption', 'Family History','Previous Heart Problems'
]

# Filtrar apenas as colunas relevantes + alvo
df_model = df[relevant_cols + ['Heart Attack Risk']].copy()

X = pd.get_dummies(df_model.drop(columns=['Heart Attack Risk']), drop_first=True)
y = df['Heart Attack Risk']

# Normalizar
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Treino e teste
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, stratify=y, random_state=42
)

# Modelo 1: Regressão Logística
logreg = LogisticRegression(max_iter=1000)
logreg.fit(X_train, y_train)
print("\n=== Regressão Logística ===")
print(classification_report(y_test, logreg.predict(X_test), zero_division=0))
print("ROC-AUC:", roc_auc_score(y_test, logreg.predict_proba(X_test)[:,1]))

# Modelo 2: Random Forest
rf = RandomForestClassifier(n_estimators=200, random_state=42)
rf.fit(X_train, y_train)
print("\n=== Random Forest ===")
print(classification_report(y_test, rf.predict(X_test), zero_division=0))
print("ROC-AUC:", roc_auc_score(y_test, rf.predict_proba(X_test)[:,1]))

# Modelo 3: KNN
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
print("\n=== KNN ===")
print(classification_report(y_test, knn.predict(X_test), zero_division=0))
print("ROC-AUC:", roc_auc_score(y_test, knn.predict_proba(X_test)[:,1]))
```
